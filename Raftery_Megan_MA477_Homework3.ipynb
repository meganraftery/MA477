```python
x1=np.linspace(-4,4,45)

y1= 3*math.e**(-0.25*x1**2)
y2 = 2.8*math.e**(-0.15*(x1-0.1)**2)
fig=plt.figure(figsize=(7,6))




axes1=fig.add_axes([-.1,0.2,1,1])
axes1.plot(x1,y1,'r-.',linewidth = 3,label = '3e^-0.25x^2',)
axes1.plot(x1,y2,label = '2.8e^-1.5(x-0.1)^2',color = 'c', marker='o',markerfacecolor='gold',markeredgecolor='black',markeredgewidth = 2,markersize=11,linestyle='-',linewidth=2)
axes1.legend(loc='upper right')

axes2=fig.add_axes([0.2,0.3,0.4,0.4])
fig.add_axes()

axes2.text(0,.8,' Cool',size=18,color='red')
axes2.text(0.25,.6,' Plot',size=18,color='blue')
axes2.text(.5,.4,' Bro',size=18,color='green')

axes1.set_title('Many Plots',fontsize=24)
axes2.set_title('Small Plot',fontsize=14)

axes1.set_xlabel('That took a while',fontsize=18)
```
